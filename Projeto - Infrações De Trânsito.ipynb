```python
#Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
infracoes = pd.read_csv('2020-03_infracoes_com_np.csv', sep=';')
infracoes
```

```python
# Criando um Plot com o NÃºmero de InfraÃ§Ãµes com base no CÃ³digo
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(infracoes["Codigo da Infracao"], color="#33cc33",kde=True, ax=ax)
ax.set_title('NÃºmero de InfraÃ§Ãµes com base no CÃ³digo Da InfraÃ§Ã£o', fontsize= 15)
plt.ylabel("Quantidade", fontsize= 15)
plt.xlabel("CÃ³digo", fontsize= 15)
plt.show()
```

```python
# Criando um Plot que mostre o nÃºmero de infraÃ§Ãµes pertencentes a cada estado
sns.set_style("whitegrid")
g = sns.catplot(y="UF", data=infracoes, kind="count", palette="Reds_r", height=7, aspect=1.5)
g.ax.set_title("InfraÃ§Ãµes por Estado",fontdict={'size':18})
g.ax.xaxis.set_label_text("NÃºmero de InfraÃ§Ãµes",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Estado",fontdict= {'size':16})
plt.show()
```
